```python
import pandas as pd
```

```python
daehyun = pd.read_csv('대현동 상가 매물.csv', encoding = 'utf-8')

daesin= pd.read_csv('대신동 상가 매물.csv', encoding = 'utf-8')

yeonhui = pd.read_csv('연희동 상가 매물.csv', encoding = 'utf-8')

changcheon = pd.read_csv('창천동 상가 매물.csv', encoding = 'utf-8')
```

```python
final = pd.concat([changcheon, yeonhui, daehyun, daesin])

final[['보증금', '월세']] = final['가격'].str.split('/', expand=True)

# '보증금' 열과 '월세' 열의 데이터 타입을 숫자로 변환 (옵션)
final['보증금'] = final['보증금'].str.replace('월세', '').str.replace(',', '')
final['월세'] = final['월세'].str.replace('월세', '').str.replace(',', '').astype(int)
final = final.drop(columns = ['가격'])
final['전용 면적'] = final['면적'].str.split('㎡').str[1].str.split('(').str[0].str.replace('/', '').astype(float)
final['평'] = final['전용 면적'] / 3.3058

def convert_to_number(s):
    # '억'을 포함하는 경우
    if '억' in s:
        parts = s.split('억')  # '억'을 기준으로 분리
        won = parts[1].replace(',', '')  # 숫자 부분에서 쉼표 제거
        if won == '' :
          return int(parts[0]) * 10000
        else :
          return int(parts[0]) * 10000 + int(won)  # 억에 대한 숫자 변환 후 합산
    else:
        return int(s.replace(',', ''))  # 쉼표 제거 후 숫자 변환

# 보증금 열의 데이터를 숫자로 변환
final['보증금'] = final['보증금'].apply(convert_to_number)

final['층'] = final['층/총층'].str.split('/').str[0]
final['구분'] = final['위치'].str[-3:]
```

```python
final= final.drop(columns = ['Unnamed: 0', '위치', '면적', '층/총층'])
final
```
